```python
df_info['Hero|Publisher'].value_counts()
```

```python
#split Hero|Publisher columns
df_info[["Hero", "Publisher"]] = df_info["Hero|Publisher"].str.split("|", expand = True)
df_info.head(2)
```

```python
#concatenate data along hero names
df = pd.merge(df_info, df_powers, left_on='Hero', right_on='hero_names')
```

```python
#drop hero/publisher
df = df.drop(columns = ["Hero|Publisher", "hero_names"])
```

```python
#uniform same colors and verify
df["Hair color"].replace({"black" : "Black", "Brownn" : "Brown", "blond" : "Blond"}, inplace = True)
df["Hair color"].value_counts()
```

```python
df['Skin color'].value_counts()
```

```python
df['Alignment'].value_counts()
```

```python
#split height and weight
df['Measurements'].value_counts()
```

```python
df[["Height in cm", "Weight in kg"]] = df["Measurements"].str.split(",", expand = True)
df.head()
```

```python
#replace units in height column
to_replace = ["{", "}", ":", "'", "Height", "cm"]
for char in to_replace:
    df["Height in cm"] = df["Height in cm"].str.replace(char,'', regex = False)
df["Height in cm"].head()
```

```python
#replace units in weight column
to_replace = ["{", "}", ":", "'", "Weight", "kg"]
for char in to_replace:
    df["Weight in kg"] = df["Weight in kg"].str.replace(char,'', regex = False)
df["Weight in kg"].head()
```

```python
#drop Measurements
df = df.drop(columns = ["Measurements"])
```

```python
df.head()
```

```python
df['Powers'].value_counts()
```

```python
df['powers_split'] = df['Powers'].str.split(',',expand = False)
exploded = df.explode('powers_split')
```

```python
#saving the unique values from the exploded column
cols_to_make = exploded['powers_split'].dropna().unique()
cols_to_make
```

```python
for col in cols_to_make:
    df[col] = df['Powers'].str.contains(col)
df.head()
```

```python
#change dtype to float to find mean
df["Height in cm"] = df["Height in cm"].astype(float)
df["Weight in kg"] = df["Weight in kg"].astype(float)
```

```python
#average weight of super powers who have Super Speed to those who do not.

super_speed = df['Super Speed']==True
no_super_speed = df['Super Speed']==False

#avg weight of super heros with super speed

avg_weight = df.groupby(super_speed)['Weight in kg'].mean()
print (avg_weight)
```

```python
#avg weight of super heros without super speed

avg_weight_no_super_speed = df.groupby(no_super_speed)['Weight in kg'].mean()
print (avg_weight_no_super_speed)
```

```python
#average height of heroes for each publisher

avg_height = df.groupby('Publisher')['Height in cm'].mean()
print (avg_height)
```
